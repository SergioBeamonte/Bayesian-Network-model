# Data Quality Agent implementation

**Drop columns with extreme missing values**

```python
missing_percent = (df.isnull().sum() / len(df)) * 100
high_missing_cols = missing_percent[missing_percent > 95].index.tolist()

print(f"Columns to drop due to high missing values: {len(high_missing_cols)}")

df = df.drop(columns=high_missing_cols)
```

**Drop identifier, leakage & high-cardinality columns**

```python
# Rule 2: Drop identifier and non-predictive columns
drop_cols = [
    'id', 'member_id', 'url', 'emp_title', 'desc', 'title',
    'zip_code'
]

# Drop only if present
drop_cols = [col for col in drop_cols if col in df.columns]
df = df.drop(columns=drop_cols)

print("Dropped identifier & high-cardinality columns")
```

**Filter only completed loans**

```python
# Rule 3: Keep only completed loan outcomes
valid_status = [
    'Fully Paid',
    'Charged Off',
    'Default',
    'Does not meet the credit policy. Status:Fully Paid',
    'Does not meet the credit policy. Status:Charged Off'
]

df = df[df['loan_status'].isin(valid_status)]

print("Remaining loan_status distribution:")
print(df['loan_status'].value_counts())
```

**Create binary target variable (Agent Decision Label)**

```python
# Rule 4: Create target variable (Accepted vs Rejected)
df['loan_accepted'] = df['loan_status'].apply(
    lambda x: 1 if 'Fully Paid' in x else 0
)

df['loan_accepted'].value_counts()
```

**Handle numeric anomalies**

```python
# Rule 5: Remove invalid numeric values

# annual_inc must be > 0
df = df[df['annual_inc'] > 0]

# dti must be between 0 and 60 (industry-safe cap)
df = df[(df['dti'] >= 0) & (df['dti'] <= 60)]

print("Dataset shape after numeric anomaly handling:", df.shape)
```

# STEP-3: Feature Selection & Encoding Agent

**Separate target & drop leakage columns**

```python
# Separate target variable
y = df['loan_accepted']

# Drop target and original status column from features
X = df.drop(columns=['loan_accepted', 'loan_status'])

print("Feature matrix shape:", X.shape)
```

**Select core, business-relevant features**

```python
core_features = [
    # Loan & pricing
    'loan_amnt', 'term', 'int_rate', 'installment',
    
    # Borrower financial health
    'annual_inc', 'dti', 'verification_status',
    'emp_length', 'home_ownership',
    
    # Credit behavior
    'fico_range_low', 'fico_range_high',
    'open_acc', 'pub_rec', 'revol_bal', 'revol_util',
    'total_acc',
    
    # Loan purpose
    'purpose', 'grade', 'sub_grade'
]

# Keep only features that exist
core_features = [col for col in core_features if col in X.columns]

X = X[core_features]

print("Selected features:", X.columns.tolist())
```

**Identify numeric vs categorical features**

```python
numeric_features = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_features = X.select_dtypes(include=['object']).columns.tolist()

numeric_features, categorical_features
```

**Handle missing values**

```python
from sklearn.impute import SimpleImputer

# Imputers
num_imputer = SimpleImputer(strategy='median')
cat_imputer = SimpleImputer(strategy='most_frequent')

X[numeric_features] = num_imputer.fit_transform(X[numeric_features])
X[categorical_features] = cat_imputer.fit_transform(X[categorical_features])
```

**Encode categorical variables**

```python
X = pd.get_dummies(X, columns=categorical_features, drop_first=True)

print("Encoded feature shape:", X.shape)
```

```python
X.isnull().sum().sum(), X.shape
```

# Step-4: Risk Prediction Model

**Train / Test split**

```python
from sklearn.model_selection import train_test_split

# Optional: sample for faster experimentation (recommended)
df_sample_X, _, df_sample_y, _ = train_test_split(
    X, y, 
    train_size=300_000, 
    stratify=y, 
    random_state=42
)

X_train, X_test, y_train, y_test = train_test_split(
    df_sample_X,
    df_sample_y,
    test_size=0.2,
    stratify=df_sample_y,
    random_state=42
)

X_train.shape, X_test.shape
```

**Feature scaling (numeric stability)**

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

**Train Logistic Regression (Baseline Model)**

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(
    max_iter=1000,
    class_weight='balanced',  # handles class imbalance
    n_jobs=-1
)

model.fit(X_train_scaled, y_train)
```

**Model evaluation**

```python
from sklearn.metrics import (
    classification_report,
    roc_auc_score
)

y_pred = model.predict(X_test_scaled)
y_proba = model.predict_proba(X_test_scaled)[:, 1]

print(classification_report(y_test, y_pred))
print("ROC-AUC:", roc_auc_score(y_test, y_proba))
```

**Interpret coefficients (Decision reasoning)**

```python
import pandas as pd
import numpy as np

feature_importance = pd.Series(
    model.coef_[0],
    index=X.columns
).sort_values()

feature_importance.head(10), feature_importance.tail(10)
```
